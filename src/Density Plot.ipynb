```python
###VALLEY SITE

fig, (ax, ax_t) = plt.subplots(1, 2, figsize=(10, 6), sharey=True, 
                               gridspec_kw={'width_ratios': [4, 1]})
plt.subplots_adjust(wspace=0) # Remove gap between plot and table

#fig, ax = plt.subplots(figsize=(8, 6)) 

#PLOTTING THE DEPTH at which density taken
depth = [78.5, 68.5, 56, 43.5, 26.5, 10]
depth_r = list(reversed(depth))

#ADDING THE DENSITY VALUES FROM DIFFERENT INSTRUMENTS 
x2 = [557, 250, 233, 264, 248, 254] #density cutter
x2_r = list(reversed(x2))

x3 = [212, 160, 181, 213, 180, 205] #denoth
x3_r = list(reversed(x3)) 


#CREATING SCATTER PLOTS 
line1 = ax.scatter(x2_r, depth_r, color='turquoise', label='Density Cutter', marker='o')
line2 = ax.scatter(x3_r, depth_r, color='magenta', label='Denoth', marker='.')
#ax.plot(x2_r, depth_r, color='turquoise', label='Density Cutter')
#ax.plot(x3_r, depth_r, color='orange', label='Denoth')


#ADD THE LAYERS 
layers = [82, 74, 62, 49, 33, 20]
# Surface layer
ax.axhline(y=82, color='C1', linestyle='-', label='Snowpit surface')

# Internal layers
layer_depths = [74, 62, 49, 33, 20]
for ld in layer_depths:
    ax.axhline(y=ld, color='C1', linestyle='--')

ax.plot([], [], color='C1', linestyle='--', label='Snowpit layers')


#Temp data for 2nd axis
depth_temp = [82, 78.5, 68.5, 56, 43.5, 26.5, 10]
temp = [-5.6, -5.3, -1.1, 0, -0.1, -0.1, 0]
ax2 = ax.twiny() # Create the second X-axis
line3, = ax2.plot(temp, depth_temp, color='red', label='Temperature', marker="x", linewidth=1.5)


#Tidying main plot
ax.set_ylim(0, 100)
ax.set_xlim(0, 600) 
ax.set_xlabel('Density (kg/m$^3$)')
ax2.set_xlabel('Temperature (°C)', color='black')
ax2.tick_params(axis='x', colors='black')

ax.set_ylabel('Depth (cm)')
ax.set_title('Comparison of Density from Density Dutter and Denoth at Valley Site')
ax.legend(loc='upper left')

lines = [line1, line2, line3]
labels = [l.get_label() for l in lines]
ax.legend(lines, labels, loc='lower left', fontsize='small')

# --- TABLE SECTION (ax_t) ---
# Table data aligned with the midpoints (depth)
hardness = ['4', '3', '4', '3', '1', '1']
snow_type = ['/', 'O', 'O ●', 'O ●', '□', '∧']

# Format the table axis
ax_t.set_xlim(0, 2)
ax_t.set_xticks([0.5, 1.5])
ax_t.set_xticklabels(['Hardness', 'Snow Type'])
ax_t.xaxis.tick_top() # Move column headers to top
ax_t.tick_params(axis='both', which='both', length=0) # Hide tick marks

# Add vertical dividers for columns
ax_t.axvline(x=0, color='black', linewidth=0.8)
ax_t.axvline(x=1, color='gray', linewidth=0.5)
ax_t.axvline(x=2, color='black', linewidth=0.8)

# Add horizontal dividers matching the layers exactly
for h in [0] + layers:
    ax_t.axhline(y=h, color='black', linewidth=0.5)

# Place text at the depth midpoints

for d, h_val, s_val in zip(depth, hardness, snow_type):
    ax_t.text(0.5, d, h_val, ha='center', va='center', fontname='DejaVu Sans')
    ax_t.text(1.5, d, s_val, ha='center', va='center', fontname='DejaVu Sans')

# Hide axis spines for a clean table look
for spine in ax_t.spines.values():
    spine.set_visible(False)



#plt.grid(True)
plt.show
plt.savefig('DensityValley')
```
