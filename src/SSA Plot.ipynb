```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#VALLEY SITE

fig, (ax, ax_t) = plt.subplots(1, 2, figsize=(10, 6), sharey=True, 
                               gridspec_kw={'width_ratios': [4, 1]})
plt.subplots_adjust(wspace=0) # Remove gap between plot and table

#PLOTTING THE DEPTH at which SSA taken (surface each layer)
depth = [82, 79, 74, 62, 49, 33, 20]
depth_r = list(reversed(depth))

#ADDING THE SSA VALUES FROM DIFFERENT INSTRUMENTS 
#Infrasnow
xI = [5.1, 8.6, 26.2, 31.8, 25.4, 24.7, 13.5]
xI_r = list(reversed(xI))

xI1 = [5.3, 11.5, 22.3, 39.8, 27.4, 26.2, 18.3]
xI1_r = list(reversed(xI1))


depth_miss = [82, 79, 74, 62, 49, 33]
depth_r_miss = list(reversed(depth_miss))
xI2 = [7.3, 10.6, 29.4, 28.3, 31.4, 22.2]
xI2_r = list(reversed(xI2))

#Icecube
x3 = [11.5, 9.9, 31, 31.5, 28.2, 13.7, 9.1]
x3_r = list(reversed(x3)) 


#CREATING SCATTER PLOTS 
line1= ax.scatter(xI_r, depth_r, color='turquoise', label='Infrasnow 1', marker='P')
line2= ax.scatter(xI1_r, depth_r, color='red', label='Infrasnow 2', marker='P')
line3= ax.scatter(xI2_r, depth_r_miss, color='green', label='Infrasnow 3', marker='P')
line4= ax.scatter(x3_r, depth_r, color='darkviolet', label='Icecube', marker='o')


#ADD THE LAYERS 
layers = [82, 74, 62, 49, 33, 20]
depth = [78.5, 68.5, 56, 43.5, 26.5, 10] #MIDPOINTS
# Surface layer
ax.axhline(y=82, color='C1', linestyle='-', label='Snowpit surface')

# Internal layers
layer_depths = [74, 62, 49, 33, 20]
for ld in layer_depths:
    ax.axhline(y=ld, color='bisque', linestyle='--')

ax.plot([], [], color='bisque', linestyle='--', label='Snowpit layers')


#Temp data for 2nd axis
depth_temp = [82, 78.5, 68.5, 56, 43.5, 26.5, 10]
temp = [-5.6, -5.3, -1.1, 0, -0.1, -0.1, 0]
ax2 = ax.twiny() # Create the second X-axis
line5, = ax2.plot(temp, depth_temp, color='red', label='Temperature', marker="x", linewidth=1.5)


#Tidying main plot
ax.set_ylim(0, 100)
ax.set_xlim(0, 50) 
ax.set_xlabel('SSA (1/mm)')
ax2.set_xlabel('Temperature (°C)', color='black')
ax2.tick_params(axis='x', colors='black')

ax.set_ylabel('Depth (cm)')
ax.set_title('Comparison of SSA from IceCube and InfraSnow at Valley Site')
ax.legend(loc='upper left')

lines = [line1, line2, line3, line4, line5]
labels = [l.get_label() for l in lines]
ax.legend(lines, labels, loc='upper right', fontsize='small')


# --- TABLE SECTION (ax_t) ---
# Table data aligned with the midpoints (depth)
hardness = ['4', '3', '4', '3', '1', '1']
snow_type = ['/', 'O', 'O ●', 'O ●', '□', '∧']

# Format the table axis
ax_t.set_xlim(0, 2)
ax_t.set_xticks([0.5, 1.5])
ax_t.set_xticklabels(['Hardness', 'Snow Type'])
ax_t.xaxis.tick_top() # Move column headers to top
ax_t.tick_params(axis='both', which='both', length=0) # Hide tick marks

# Add vertical dividers for columns
ax_t.axvline(x=0, color='black', linewidth=0.8)
ax_t.axvline(x=1, color='gray', linewidth=0.5)
ax_t.axvline(x=2, color='black', linewidth=0.8)

# Add horizontal dividers matching the layers exactly
for h in [0] + layers:
    ax_t.axhline(y=h, color='black', linewidth=0.5)

# Place text at the depth midpoints

for d, h_val, s_val in zip(depth, hardness, snow_type):
    ax_t.text(0.5, d, h_val, ha='center', va='center', fontname='DejaVu Sans')
    ax_t.text(1.5, d, s_val, ha='center', va='center', fontname='DejaVu Sans')

# Hide axis spines for a clean table look
for spine in ax_t.spines.values():
    spine.set_visible(False)



#ADD GRAIN TYPES AS ANNOTATIONS 
#ax.annotate('Hardness and Snow Type', xy=(350, 350), xycoords='figure points') 
#ax.annotate('o', xy=(450, 290), xycoords='figure points') 
##ax.annotate('/', xy=(450, 325), xycoords='figure points') 
#ax.annotate(".", xy=(450, 250), xycoords='figure points') 
#ax.annotate(".", xy=(450, 200), xycoords='figure points') 
#ax.annotate('[]', xy=(450, 140), xycoords='figure points') 
#ax.annotate('^', xy=(450, 90), xycoords='figure points') 

#ADD HARDNESS
#ax.annotate('4', xy=(420, 325), xycoords='figure points') 
#ax.annotate('3', xy=(420, 290), xycoords='figure points') 
#ax.annotate('4', xy=(420, 250), xycoords='figure points') 
#ax.annotate('3', xy=(420, 200), xycoords='figure points') 
#ax.annotate('1', xy=(420, 140), xycoords='figure points') 
#ax.annotate('1', xy=(420, 90), xycoords='figure points')


#ax.set_ylim(0, 90)
#ax.set_yticks([0, 20, 33, 49, 62, 74, 82])
#ADD THE MELT LAYERS
#plt.hlines(y=[75, 63, 50, 38, 34], xmin=0, xmax= 50, colors=['b'], linestyles=[':'], label= 'Melt layers',)


plt.show
plt.savefig('SSAValley')
```

```python
#MOUNTAIN SITE

fig, (ax, ax_t) = plt.subplots(1, 2, figsize=(10, 6), sharey=True, 
                               gridspec_kw={'width_ratios': [4, 1]})
plt.subplots_adjust(wspace=0) # Remove gap between plot and table


#PLOTTING THE DEPTH at which the SSA taken
depth = [175, 172, 125, 91, 78, 59, 14]
depth_r = list(reversed(depth))

midpoints = [148.5, 108, 84.5, 69.5, 36.5, 7]

#ADDING THE SSA VALUES FROM DIFFERENT INSTRUMENTS 
#Infrasnow
xI = [6.9, 13.7, 34.7, 23, 17.9, 20.8, 13.4]
xI_r = list(reversed(xI))

xI1 = [23.4, 18.4, 31.1, 24.5, 16, 16.3, 11]
xI1_r = list(reversed(xI1))

xI2 = [11.6, 18.6, 33.6, 25.6, 19.1, 23.3, 11]
xI2_r = list(reversed(xI2))

#Icecube
x3 = [28.5, 36.3, 33.1, 27.7, 15.2, 12.3, 8.4]
x3_r = list(reversed(x3)) 


#CREATING SCATTER PLOTS 
line1= ax.scatter(xI_r, depth_r, color='turquoise', label='Infrasnow 1', marker='P')
line2= ax.scatter(xI1_r, depth_r, color='red', label='Infrasnow 2', marker='P')
line3= ax.scatter(xI2_r, depth_r, color='green', label='Infrasnow 3', marker='P')
line4= ax.scatter(x3_r, depth_r, color='darkviolet', label='Icecube', marker='o')

#ADD THE LAYERS 
layers = [175, 125, 91, 78, 59, 14]
# Surface layer
ax.axhline(y=175, color='C1', linestyle='-', label='Snowpit surface')

# Internal layers
layer_depths = [125, 91, 78, 59, 14]
for ld in layer_depths:
    ax.axhline(y=ld, color='bisque', linestyle='--')

ax.plot([], [], color='bisque', linestyle='--', label='Snowpit layers')

#Temp data for 2nd axis
depth_temp = [175, 148.5, 108, 84.5, 69.5, 36.5, 7]
temp = [-1.5, -6.6, -4, -3.1, -2.8, -1.7, -0.3]
ax2 = ax.twiny() # Create the second X-axis
line5, = ax2.plot(temp, depth_temp, color='red', label='Temperature', marker="x", linewidth=1.5)


#Tidying main plot
ax.set_ylim(0, 190)
ax.set_xlim(0, 50) 
ax.set_xlabel('SSA (1/mm)')
ax.set_ylabel('Depth (cm)')
ax.set_title('Comparison of SSA from IceCube and InfraSnow at Mountain Site')

ax2.set_xlabel('Temperature (°C)', color='black')
ax2.tick_params(axis='x', colors='black')

lines = [line1, line2, line3, line4, line5]
labels = [l.get_label() for l in lines]
ax.legend(lines, labels, loc='upper right', fontsize='small')

# --- TABLE SECTION (ax_t) ---
# Table data aligned with the midpoints (depth)
hardness = ['2', '3', '3', '2', '1', '2']
snow_type = ['●', '●', '●', '□', '□', '∧']

# Format the table axis
ax_t.set_xlim(0, 2)
ax_t.set_xticks([0.5, 1.5])
ax_t.set_xticklabels(['Hardness', 'Snow Type'])
ax_t.xaxis.tick_top() # Move column headers to top
ax_t.tick_params(axis='both', which='both', length=0) # Hide tick marks

# Add vertical dividers for columns
ax_t.axvline(x=0, color='black', linewidth=0.8)
ax_t.axvline(x=1, color='gray', linewidth=0.5)
ax_t.axvline(x=2, color='black', linewidth=0.8)

# Add horizontal dividers matching the layers exactly
for h in [0] + layers:
    ax_t.axhline(y=h, color='black', linewidth=0.5)

# Place text at the depth midpoints
# depth is [148.5, 108, 84.5, 69.5, 36.5, 7]
for d, h_val, s_val in zip(midpoints, hardness, snow_type):
    ax_t.text(0.5, d, h_val, ha='center', va='center', fontname='DejaVu Sans')
    ax_t.text(1.5, d, s_val, ha='center', va='center', fontname='DejaVu Sans')

# Hide axis spines for a clean table look
for spine in ax_t.spines.values():
    spine.set_visible(False)


plt.show
plt.savefig('SSAMountain')
```
